# Load all packages and start browser

```python
import mechanicalsoup
import os
browser = mechanicalsoup.StatefulBrowser()
```

# Specify the folders to download from pokemon showdown

```python
showdown_url = "https://play.pokemonshowdown.com/sprites/"

###
# NOTE: Already downloaded folders


###
# TODO: ADD links to the following list to start automatic downloads
###


all_gens_url = ["xy"]
```

# Download all the sprites to the "sprites" folder

```python
##REAL CODE; we want to do test code
for gen_url in all_gens_url:
    browser.open(showdown_url+gen_url)
    
    ###
    # Following creates a folder for each generation, if the folder doesn't exist
    ###
    download_path = os.path.join("sprites",gen_url)
    full_folder_path = os.path.join(os.getcwd(),download_path)
    print("Analyzing folder :: "+full_folder_path)
    print("Has the folder been created? : ",os.path.exists(full_folder_path))
    if(not os.path.exists(full_folder_path)):
        try:
            os.mkdir(full_folder_path)
        except OSError:
            print("Creation of the direction %s failed" %full_folder_path)
        else:
            print("Successfully created the directory %s" %full_folder_path)
    print("*************************************************************************")
    
    ###
    # Obtain the links and download the images
    # NOTE: ONLY DOWNLOADS PNG files at the moment
    ###
    
    #Note: MechanicalSoup basically makes beautifulsoup easier to use
    #Have to use both documentations to understand wtf is happening lmao
    all_links = browser.links('png')
    
    for link in all_links:
        #'href' in links.attrs dictionary tells us the link to follow, and thus the filename
        if('href' in link.attrs):
            print("Downloading :: "+link.get('href'))
            browser.download_link(link=link, file=os.path.join(download_path,link.get('href')))
        else:
            print(link + " doesn't have an href attribute. Can't save image without file name!")
```
