```python
import os
import json
from pprint import pprint
```

```python
sprites_dir = os.path.join(os.getcwd(), "sprites")
data_dir = os.path.join(os.getcwd(), "data")
```

```python
#below loads "pokedex.json" to variable data
#which is a python dictionary
with open('from_js_pokedex.json') as f:
    data = json.load(f)
    
newDict = data.copy()
```

```python
for pokemon in data:
    
    if("otherFormes" in data[pokemon]):
        
        # otherformes (note the e) are Mega's and Alolan variants
        for otherForm in data[pokemon]["otherFormes"]:
            print(otherForm)
            if not otherForm in data:
                print(">>>>>"+otherForm)
                newDict[otherForm] = data[pokemon]
                
                
    if("otherForms" in data[pokemon]):
        
        # otherforms (*NO* e) are species variants
        for otherForm in data[pokemon]["otherForms"]:
            print(otherForm)
            if not otherForm in data:
                print(">>>>"+otherForm)
                newDict[otherForm] = data[pokemon]

```
